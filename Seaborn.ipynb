```python
import seaborn as sns
```

```python
tip=sns.load_dataset('tips')
```

```python
tip.head()
```

```python
%matplotlib inline
```

```python
sns.distplot(tip['tip'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='kde')
```

```python
sns.pairplot(tip,hue='sex')
```

```python
sns.rugplot(tip['total_bill'])
```

```python
sns.barplot(x='day',y='total_bill',data=tip)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tip)
```

```python
sns.boxplot(x='day',y='total_bill',data=tip,)
```

```python
sns.boxplot(x='sex',y='tip',data=tip,hue='smoker')
```

```python
sns.distplot(tip['tip'],kde=False)
```

```python
sns.violinplot(x='day',y='total_bill',data=tip,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tip)
```

```python
sns.factorplot(x='day',y='total_bill',data=tip)
```
